```python
import sympy as sp
import einsteinpy.symbolic as gr
sp.init_printing()
```

```python
t, r, theta, phi = sp.symbols('t, r, theta, phi')
u, a, c = sp.symbols('mu, a, c')
rhox, Dx, Ex = sp.symbols('rho, Delta, Sigma')

rho = sp.sqrt(r**2 + a**2*(sp.cos(theta))**2)
D = r**2 - 2*u*r + a**2
E = sp.sqrt((r**2 + a**2)**2 - a**2*D*(sp.sin(theta))**2)

n = 4
```

## MÃ©trica

```python
metric = [[0 for i in range(4)] for i in range(4)]

metric[0][0] = sp.simplify((D - a**2*(sp.sin(theta))**2)/rho**2)
metric[0][3] = sp.simplify(2*u*a*r*(sp.sin(theta))**2/rho**2)
metric[1][1] = -rho**2/D
metric[2][2] = -rho**2
metric[3][0] = sp.simplify(2*u*a*r*(sp.sin(theta))**2/rho**2)
metric[3][3] = sp.simplify(-E**2*(sp.sin(theta))**2/rho**2)

coord = [t, r, theta, phi]
metric_gr = gr.MetricTensor(metric, coord)

metric_gr.tensor()
```

```python
metricx = metric_gr.subs([(r**2 + a**2*(sp.cos(theta))**2, rhox**2), (r**2 - 2*u*r + a**2, Dx)])
metricx = metricx.subs((r**2 + a**2)**2 - a**2*Dx*(sp.sin(theta))**2, Ex**2)


metricx.tensor()
```

## Christoffel

```python
chris = gr.ChristoffelSymbols.from_metric(metric_gr)
```

```python
for i in range(n):
    for j in range(n):
        for k in range(n):
            if j != k:
                if chris[i, j, k] != 0:
                    chrisx = sp.simplify(chris[i, j, k].subs([(r**2 + a**2*(sp.cos(theta))**2, rhox), (r**2 - 2*u*r + a**2, Dx)]))
                    chrisx = chrisx.subs((r**2 + a**2)**2 - a**2*Dx*(sp.sin(theta))**2, Ex)
                   # chrisx = 
                    chrisx = chrisx.ratsimp().cancel()
                    chrisx = chrisx.factor()

                    display([i, j, k], chrisx)
```

# Ricci
